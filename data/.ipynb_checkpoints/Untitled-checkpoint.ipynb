```python
import pandas as pd
```

```python
df = pd.read_csv("deaths_per_season.csv", header=0)
```

```python
df.head()
```

```python
states_df = df.iloc[:, 6:7]
```

```python
states_df.head()
```

```python
df = df.iloc[:, 11:]
df.head()
```

```python
result_df = pd.concat([states_df, df], axis=1)
result_df.head()
```

```python
grouped_df = result_df.groupby(['Province_State']).sum()
```

```python
grouped_df.head()
```

```python
grouped_df.to_csv("Covid_per_season_per_state.csv")
```

```python
calc_df = grouped_df.copy()
```

```python
calc_df.shape
```

```python
calc_df.columns[1] + '_rate'
```

```python
for i in range(1, calc_df.shape[1]):
    new_col_name = calc_df.columns[i] + '_rate'
    calc_df[new_col_name] = calc_df.iloc[:, i] / calc_df.iloc[:, 0]
```

```python
calc_df.head()
```

```python
calc_df.head()
```

```python
col_names = ['spring', 'summer', 'autumn', 'winter']
for i in range(4):
    calc_df[col_names[i]] = calc_df.iloc[:, [col + i + 1 for col in spring_col]].sum(axis=1) - \
    calc_df.iloc[:, [col + i for col in spring_col]].sum(axis=1)
calc_df.head()
```

```python
spring_col = [1, 5, 9]
summer_col = [2, 6, 10]
autumn_col = [3, 7, 11]
winter_col = [4, 8, 12]
col_names = ['spring', 'summer', 'autumn', 'winter']
col_rate_names = ['spring_rate', 'summer_rate', 'autumn_rate', 'winter_rate']
for i in range(4):
    calc_df[col_rate_names[i]] = calc_df[col_names[i]] / calc_df.iloc[:, 0]
```
